## Calculate Urban Heat Anomalies
We can calculate the tree fraction, grass fraction, etc. by dividing the tree areas by the area column, which is for area of the census tract.

```python
data['CUHI_day_summer'] = data['AT_urb_day_CT_act_summer'] - data['AT_rur_day_summer']
data['CUHI_night_summer'] = data['AT_urb_night_CT_act_summer'] - data['AT_rur_night_summer']
data['SUHI_day_summer'] = data['LST_urb_day_CT_act_summer'] - data['LST_rur_day_summer']
data['SUHI_night_summer'] = data['LST_urb_night_CT_act_summer'] - data['LST_rur_night_summer']
```

```python
cols = ['ESABuilt', 'ESAGrass', 'ESATree']
for col in cols:
    data[col+'_Area'] = data[col] / data['Area']
```

## Add Median Income Data

```python
income_df = pd.read_csv('data/ACSDT5Y2017.B20018_2025-02-04T152116/ACSDT5Y2017.B20018-Data.csv', skiprows=1)
income_df['Census_geoid'] = income_df['Geography'].str[9:]
income_df['Census_geoid'] = income_df['Census_geoid'].astype('int64')
income_df.rename(columns={'Estimate!!Median earnings in the past 12 months (in 2017 inflation-adjusted dollars)': 'Median Income'}, inplace=True)

data = data.merge(income_df[['Census_geoid', 'Median Income']], how='left', on='Census_geoid')
```

```python
data.loc[data['Median Income'] == '-', 'Median Income'] = np.nan

data['Median Income'] = data['Median Income'].str.replace('+', '', regex=True)
data['Median Income'] = data['Median Income'].str.replace('-', '', regex=True)
data['Median Income'] = data['Median Income'].str.replace(',', '', regex=True)

# Fill null values with a temporary placeholder
data['Median Income'] = data['Median Income'].fillna(-999)

# Convert to integer
data['Median Income'] = data['Median Income'].astype('int64')

# Set the placeholder back to nulls
data['Median Income'] = data['Median Income'].replace(-999, np.nan)
```

## Prepare the Shapefile

```python
import os
os.environ['USE_PYGEOS'] = '0'
import geopandas as gpd
```

```python
gdf = gpd.read_file('data/usa/usa.shp')
```

```python
gdf['SF'].unique()
```

```python
gdf_contiguous = gdf[['GEOID10', 'SF', 'CF', 'geometry']]
exclude = ['Alaska', 'Hawaii', 'Puerto Rico', 'American Samoa', 
           'Guam', 'Northern Mariana Islands', 'Virgin Islands']
gdf_contiguous = gdf_contiguous[~gdf_contiguous['SF'].isin(exclude)]
gdf_contiguous['GEOID10'] = gdf_contiguous['GEOID10'].astype(int)
```

```python
len(gdf_contiguous)
```

```python
gdf_contiguous.head()
```

```python
gdf_contiguous.SF.unique()
```

```python
# Shapefile columns names cannot exceed 10 characters
columns = {'AT_rur_day_summer': 'ATrurday', 'AT_rur_night_summer': 'ATrurnight',
           'AT_urb_day_CT_act_summer': 'ATurbday', 'AT_urb_night_CT_act_summer':'ATurbnight',
           'Albedo_Built': 'AlbedBuilt', 'Albedo_Grass': 'AlbedGrass',
           'Albedo_Tree': 'AlbedoTree', 'Albedo':'Albedo', 'Census_geoid': 'GEOID10',
           'ESABuilt': 'ESABuilt', 'ESAGrass': 'ESAGrass', 'ESATree': 'ESATree',
           'LST_rur_day_summer': 'LSTrurday', 'LST_rur_night_summer': 'LSTrurnigh',
           'LST_urb_day_CT_act_summer': 'LSTurbday', 'LST_urb_night_CT_act_summer': 'LSTurbnigh',
           'Urban_geoid': 'Urban_geoid', 'CUHI_day_summer': 'CUHIday', 'CUHI_night_summer':'CUHInight',
           'SUHI_day_summer':'SUHIday', 'SUHI_night_summer':'SUHInight', 
           'ESABuilt_Area': 'BuiltArea', 'ESAGrass_Area': 'GrassArea', 'ESATree_Area': 'TreeArea',
           'DEM_urb_CT_act': 'DEMurb'}
```

```python
drop_cols = ['system:index', 'AT_urb_day_all_summer', 'AT_urb_night_all_summer', 
             'Albedo_all', 'Albedo_Tree_all', 'Albedo_Built_all', 'Albedo_Grass_all',
             'DEM_rur', 'DEM_urb_CT', 'DEM_urb_all', 'ESABuilt_all', 'ESAGrass_all', 'ESATree_all',
             'LST_urb_day_all_summer', 'LST_urb_night_all_summer', 'Urban_name_extended']
data10 = data.drop(columns=drop_cols)
data10.rename(columns=columns, inplace=True)
```

```python
shape_data = gdf_contiguous.merge(data10, how='right', on='GEOID10')
```
